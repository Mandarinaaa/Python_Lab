## Бинарный поиск   O(log(n))
### Поиск

```python
import time
startTime = time.time()
def binary_search(array, key):
    print(array)
    lower_bound = 0
    upper_bound = len(array) - 1

    while lower_bound <= upper_bound:
        center = (upper_bound + lower_bound) // 2

        if array[center] > key:
            upper_bound = center - 1

        elif array[center] < key:
            lower_bound = center + 1

        elif array[center] == key:
            return center

    return -1, lower_bound

print(binary_search(mass,17))
endTime = time.time()
totalTime = endTime - startTime

print("Время:", round(endTime-startTime,6), 'seconds')
```

### Удаление

```python
def binary_delete(array, key):
    element = (binary_search(array, key))
    #print(array)
    if isinstance(element, tuple):
        return array
    del array[element]
    return (array)
    
print(binary_delete(mass.copy(), 17))
```

```python
def binary_paste(array, key):
    element = (binary_search(array, key))
    print(array)
    if isinstance(element, tuple):
        array.insert(element[1], key)
    else:
        array.insert(element, key)
    return (array)
    
print(binary_paste(mass.copy(), 1))
```

```python
class Node:
    def __init__(self, x, parent):
        self.left = None
        self.right = None
        self.data = x
        self.parent = parent
class Tree:

    def __init__(self):
        self.root = None

    def insert(self, x, node=None):
        if node is None:
            node = self.root
            if self.root is None:
                self.root = Node(x, None)
                return

        if x < node.data:
            if node.left is None:
                node.left = Node(x, node)
                return
            self.insert(x, node.left)
        else:
            if node.right is None:
                node.right = Node(x, node)
                return
            self.insert(x, node.right)

    def search(self, x, node=None):
        if node is None:
            node = self.root

        if node.data == x:
            return node

        if x < node.data:
            if node.left is None:
                return False
            return self.search(x, node.left)
        else:
            if node.right is None:
                return False
            return self.search(x, node.right)

    def print(self, node=None):
        if node is None:
            node = self.root

        if node.left:
            self.print(node.left)
        print(node.data)

        if node.right:
            self.print(node.right)

    def change_children(self, node, new_node):
        if node.parent is None:
            if new_node:
                new_node.parent = None
            self.root = new_node
            if node.left is new_node:
                self.root.right = node.right
            else:
                self.root.left = node.left
            return
        if node.parent.left is node:
            if new_node:
                new_node.parent = node.parent
                if node.left is new_node:
                    new_node.right = node.right
                else:
                    new_node.left = node.left
            node.parent.left = new_node
        else:
            if new_node:
                if node.left is new_node:
                    new_node.right = node.right
                else:
                    new_node.left = node.left
                new_node.parent = node.parent
            node.parent.right = new_node


    def rebalance(self, node):
        if node.left:
            if node.left.right:
                node.data = node.left.right.data
                node.left.right.data = None
                self.rebalance(node.left.right)
                return
            self.change_children(node, node.left)
            return
        if node.right:
            self.change_children(node, node.right)
            return
        self.change_children(node, None)

    def delete(self, x):
        node = self.search(x)
        if not node:
            return
        node.data = None
        self.rebalance(node)


T = Tree()
T.insert(5)
T.insert(1)
T.insert(8)
T.insert(16)
T.insert(4)
T.insert(3)
T.insert(7)
T.insert(8)
T.insert(9)

T.delete(5)
T.delete(8)

T.print()
```

## Фибоначчи O(ln(n))

### Поиск

```python
import time
startTime = time.time()   
class FibonacciSearch:
    def __init__(self):
        self.i = 0
        self.p = 0
        self.q = 0
        self.stop = False

    def get_fibonacci_number(self, k):
        first_number = 0
        second_number = 1
        n = 0
        while n<k:
            temp_number = second_number
            second_number = first_number + second_number
            first_number = temp_number
            n = n+1
        return first_number

    def start_init(self, massiv):
        self.stop = False
        k = 0
        n = len(massiv)
        while (self.get_fibonacci_number(k+1) < len(massiv)):
            k = k+1
        m = self.get_fibonacci_number(k+1)-(n+1)
        self.i = self.get_fibonacci_number(k) - m
        self.p = self.get_fibonacci_number(k-1)
        self.q = self.get_fibonacci_number(k-2)

    def up_index(self):
        if self.p == 1:
            self.stop = True
        self.i = self.i + self.q
        self.p = self.p - self.q
        self.q = self.q - self.p

    def down_index(self):
        if self.q == 0:
            self.stop = True
        self.i = self.i - self.q
        temp = self.q
        self.q = self.p - self.q
        self.p = temp

    def search(self, massiv, element):
        self.start_init(massiv)
        result_index = -1
        while not self.stop:
            if self.i < 0:
                self.up_index()
            elif self.i >= len(massiv):
                self.down_index()
            elif massiv[self.i] == element:
                result_index = self.i
                break
            elif element < massiv[self.i]:
                self.down_index()
            elif element > massiv[self.i]:
                self.up_index()
        if result_index == -1:
            return result_index, self.i
        return result_index

fs = FibonacciSearch()
print (mass)
print(fs.search(mass,17))

endTime = time.time()
totalTime = endTime - startTime
print("Время:", round(endTime-startTime,6), 'seconds')
```

## Интерполяционный
### Поиск

```python
import time
startTime = time.time()
def interpolation_search(massiv, key):
    print (massiv)
    low = 0
    high = len(massiv)-1
    
    while (low<=high and key >= massiv[low] and key<= massiv[high]):
        middle = low + int(((high-low)/(massiv[high]-massiv[low]))*(key-massiv[low]))
        if massiv[middle] == key:
            return middle
        if massiv[middle] < key:
            low = middle + 1
        else:
            high = middle - 1
    return -1, high

print(interpolation_search(mass,17))

endTime = time.time()
totalTime = endTime - startTime
print("Время:", round(endTime-startTime,6), 'seconds')
```

```python
global queen
global number
queen = [None] * 8
number = 0


def print_table():
    global number
    number += 1
    print('\n%d решение \t' % number)
    for x in range(8):
        for y in range(8):
            if x == queen[y]:
                print('<q>', end=' ')
            else:
                print('<->', end=' ')
        print('\t')


def attack(row, col):
    global queen
    for i in range(col):
        offset_col = abs(i - col)
        offset_row = abs(queen[i] - row)
        if queen[i] == row or offset_row == offset_col:
            return True
    return False


def decide_position(value):
    global queen
    for i in range(8):
        if not attack(i, value):
            queen[value] = i
            if value == 7:
                print_table()
            else:
                decide_position(value + 1)


decide_position(0)
```
